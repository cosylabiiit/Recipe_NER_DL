```python
train = pd.read_csv('ar_gk_train.tsv',sep = '\t')
test = pd.read_csv('ar_gk_test.tsv',sep = '\t')
```

```python
train.columns=['token','tag']
test.columns=['token','tag']
```

```python
token_list = train.token.values.tolist()
tag_list= train.tag.values.tolist()
```

```python
train.columns=['token','tag']
token_list = train.token.values.tolist()
tag_list= train.tag.values.tolist()
inputs=['']*len(token_list)
outputs=['']*len(token_list)
count=0
for i in token_list:
    if type(i) != float: 
        inputs[count]+=i
        inputs[count]+=' '
    else:
        temp=inputs[count].strip()
        inputs[count]=temp
        count+=1
count=0
for i in tag_list:
    if type(i) != float: 
        outputs[count]+=i
        outputs[count]+=' '
    else:
        temp=outputs[count].strip()
        outputs[count]=temp
        count+=1
        
only_inputs=[]
for i in inputs:
    if len(i) !=0:
        only_inputs.append(i)
        
only_outputs=[]
for i in outputs:
    if len(i) !=0:
        only_outputs.append(i)
        
import csv
with open('train.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(zip(only_inputs, only_outputs))
```

```python
test.columns=['token','tag']
token_list = test.token.values.tolist()
tag_list= test.tag.values.tolist()
inputs=['']*len(token_list)
outputs=['']*len(token_list)
count=0
for i in token_list:
    if type(i) != float: 
        inputs[count]+=i
        inputs[count]+=' '
    else:
        temp=inputs[count].strip()
        inputs[count]=temp
        count+=1
count=0
for i in tag_list:
    if type(i) != float: 
        outputs[count]+=i
        outputs[count]+=' '
    else:
        temp=outputs[count].strip()
        outputs[count]=temp
        count+=1
        
only_inputs=[]
for i in inputs:
    if len(i) !=0:
        only_inputs.append(i)
        
only_outputs=[]
for i in outputs:
    if len(i) !=0:
        only_outputs.append(i)
        
import csv
with open('test.csv', 'w') as f:
    writer = csv.writer(f)
    print(writer)
    writer.writerows(zip(only_inputs, only_outputs))
```
